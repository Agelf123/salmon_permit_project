```python
years = []
for i in range(1955,2020):
    try:
        for j in ids:
            content = requests.get(f'http://www.adfg.alaska.gov/sf/FishCounts/index.cfm?ADFG=export.JSON&countLocationID={j}&year={i}&speciesID=420')
            print(content.status_code)
            print(j)
            year = json.loads(content.content)
            df = pd.DataFrame.from_dict(year['DATA'])
            years.append(df)
    except NameError:
        print("year not available")
        
        
river = pd.concat(years, sort=False)
```

```python
river.to_csv("river.csv")
```

```python
BB_recent = pd.concat([BB_2019,BB_2018,BB_2017,BB_2016,BB_2015,BB_2014,BB_2013,BB_2012])
BB_recent.rename(columns={'Sockeye Count':'FISHCOUNT','Stream Name':'COUNTLOCATION','Date':'COUNTDATE'}, inplace=True)
BB_recent.set_index('COUNTDATE', inplace=True)
BB_recent.head()
```

```python
fish_counts = pd.concat([river,BB_recent])
```

```python
fish_counts['COUNTLOCATION']
```

```python
fish_counts.drop_duplicates(inplace=True)
fish_counts['FISHCOUNT'] = pd.to_numeric(fish_counts['FISHCOUNT'])
```

```python
fish_counts.index = pd.to_datetime(fish_counts.index, format='%Y/%m/%d')
```

```python
fish_counts = pd.pivot_table(fish_counts,index=fish_counts.index, columns = 'COUNTLOCATION')
```

```python
fish_counts = fish_counts.loc[:,~fish_counts.columns.duplicated()]
```

```python
fish_counts = fish_counts.resample('M').sum()
fish_counts.shape
```

```python
mask = (fish_counts.iloc[:, 1:] == 0).sum(1) >= len(fish_counts.columns[1:]) / 2
fish_counts = fish_counts[~mask]
```

```python
import requests,json
from bs4 import BeautifulSoup
import pandas as pd

url =  "https://www.cfec.state.ak.us/pmtvalue/X_S03T.HTM"
res = requests.get(url)
soup = BeautifulSoup(res.content,'lxml')
table = soup.find("table",{"class":"table"})

content = []

for row in table.find_all("tr"):
    row_data = []
    for col in row.find_all(["td","th"]):
        row_data.append(col.text)
    # add to content
    content.append(row_data)
    
# Format to dataframe
prices = pd.DataFrame(content[1:], columns=content[0])
prices.head()
```
