These are the rivers that I want to gather data for. They have a corresponding 2 digit code that goes with each river in order to get their fish counts.

```python
years = []
for i in range(1955,2020):
    try:
        for j in ids:
            content = requests.get(f'http://www.adfg.alaska.gov/sf/FishCounts/index.cfm?ADFG=export.JSON&countLocationID={j}&year={i}&speciesID=420')
            print(content.status_code)
            print(j)
            year = json.loads(content.content)
            df = pd.DataFrame.from_dict(year['DATA'])
            years.append(df)
    except NameError:
        print("year not available")
        
        
river = pd.concat(years, sort=False)
```

This function gets every year available for a river and adds it to a dataframe. Be aware that since it does each river separately,this function will run for over 30 minutes. In order to expedite it, it saves itself and you can access it everytime you need it after running it only one time.

```python
url =  "https://www.cfec.state.ak.us/pmtvalue/X_S03T.HTM"
res = requests.get(url,verify=False)
soup = BeautifulSoup(res.content,'lxml')
table = soup.find("table",{"class":"table"})

content = []

for row in table.find_all("tr"):
    row_data = []
    for col in row.find_all(["td","th"]):
        row_data.append(col.text)
    # add to content
    content.append(row_data)
    
# Format to dataframe
prices = pd.DataFrame(content[1:], columns=content[0])
prices.head()
```

Now we want to scraper to go the section of the Alaska Dept of F & G and get the values for Bristol Bay Drift Permits for all the years it has available. This turns out to be 1984-2019. 

```python
prices = prices[['Year','Month','Estimated Value']]# get the columns we want
```

```python
prices = prices[prices['Month'] == 'ALL']# erase all columns
prices = prices[['Year','Estimated Value']]


```

```python
prices.set_index('Year', inplace=True)
```

```python
prices.index = pd.to_datetime(prices.index, format='%Y/%m/%d')
```

```python
prices['Estimated Value']=(prices['Estimated Value'].replace( '[\$,)]','', regex=True ))

prices['Estimated Value'] = prices['Estimated Value'].astype('float64')
```

```python
prices = prices.resample('W').mean()

prices = prices.interpolate(method='spline', order=2)
```

```python
prices.head()
prices.to_csv("permit_price.csv")
```
