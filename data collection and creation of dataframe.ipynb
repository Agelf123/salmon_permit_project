```python
import requests,json
from bs4 import BeautifulSoup
import pandas as pd

url =  "https://www.cfec.state.ak.us/pmtvalue/X_S03T.HTM"
res = requests.get(url)
soup = BeautifulSoup(res.content,'lxml')
table = soup.find("table",{"class":"table"})

content = []

for row in table.find_all("tr"):
    row_data = []
    for col in row.find_all(["td","th"]):
        row_data.append(col.text)
    # add to content
    content.append(row_data)
    
# Format to dataframe
prices = pd.DataFrame(content[1:], columns=content[0])
prices.head()
```

```python
prices = prices[['Year','Month','Estimated Value']]
```

```python
prices = prices[prices['Month'] == 'ALL']
prices = prices[['Year','Estimated Value']]


```

```python
prices.set_index('Year', inplace=True)
prices.head()
```

```python
prices.index = pd.to_datetime(prices.index, format='%Y/%m/%d')
```

```python
prices['Estimated Value']=(prices['Estimated Value'].replace( '[\$,)]','', regex=True ))

prices['Estimated Value'] = prices['Estimated Value'].astype('float64')
```

```python
prices.head()
```

```python
prices = prices.resample('W').mean()

prices = prices.interpolate(method='spline', order=2)
```

```python
prices.head()
```

```python
df = pd.merge(fish_counts, prices, left_index=True, right_index=True)
```
